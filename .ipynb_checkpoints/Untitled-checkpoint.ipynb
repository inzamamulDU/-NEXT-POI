```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('Gowalla_totalCheckins.csv', sep='\t', header=None)
df.columns = ['userid','timestamp','latitude','longitude','POIid']
df.head()
```

```python
df = df[['userid','POIid','latitude','longitude','timestamp']]
df
```

```python
df.to_csv('userlocation.csv')
```

```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
geolocator = Nominatim(user_agent="myGeocoder", timeout=10)
```

```python
cityID = []   # create empty lists


for index, row in df.iterrows(): # iterate over rows in dataframe

    
    #tqdm.pandas()
    latitude = row['latitude']
    longitude = row['longitude']       
    query = str(latitude)+','+str(longitude)
    try:
        location = geolocator.reverse(query)   
        address = location.raw['address']
        city = address.get('city', '')
        print (query,city)

        cityID.append(city)
    #list_long.append(long)
    except GeocoderTimedOut as e:
        cityID.append(city)
        
    


# create new columns from lists    

df['cityID'] = cityID   
```
